В терминологии библиотеки spacy, на выходе подключения языковой модели к тексту получается документ. Т.е. spacy именует текст с языковой моделью как документ - подобно как pandas именует загруженные данные как dataframe.

Документ в терминологии библиотеки spacy - совокупность и последовательность токенов, составляющих этот документ.

Токены - это единицы текста. Токенизация - процесс и способ разбиения документа на единицы текста. Под единицами имеются в виду: фразы, предложения, абзацы, слова.

Тип данных документа после применения scapy - spacy.tokens.doc.Doc .

Для того, чтобы корректно обращаться к документу с данным типом, применяйте списки (list) и циклы (for ... in ).

Во время обработки spaCy сначала токенизирует текст, т. е. сегментирует его на слова, знаки препинания и т. д. Это делается путем применения правил, специфичных для каждого языка.


Один из примеров обращения к элементам документа средствами библиотеки spacy:

```python
analyzing_part = []
for element in doc:
        analyzing_part.append([word.spacy_function for word in element])

# где spacy_function - некоторая определенная функция из библиотеки spacy
```

#### 5. Выполните токенизацию текста

Выполните токенизацию на предложения, применив функцию .sents.

#### 7. Определите леммы слов каждого предложения

Соответственно, категории могут включать имена людей, организаций, местоположений, выражения времени, количества, медицинские коды, денежные значения и проценты и т. д.

#### 11. Поиск часто употребляемых в тексте слов

По умолчанию spacy не имеет предопределённой функции по поиску наиболее частых слов в зависимости от токенов и лемм.
Реализуем данную функциональность

Импортируем счетчик python:

https://docs.python.org/3/library/collections.html

```python
from collections import Counter
```

В зависимости от того, как вы привыкли работать, вы можете продолжить работать с dataframe или продолжить работать с документом.

```python
Необходимы леммы всего документа. Пример простого получения лемм документа:
```

Как вы видите счетчик посчитал все слова и символы, чаще всего встречаются знаки препинания, предлоги и т.д.. Для того, чтобы этого избежать нужно очистить документ от несущественных слов и знаком препинания. Данная процедура именуется как очистка от stop word : несущественные слова, не имеющие большой ценности в помощи вычислительным машинам в ответах на запросы. Конкретные стоп-слова могут меняться в зависимости от контекста.
примеры очистки от: междометий, личных местоимений, предлогов, артиклей, знаков препинания. И т.д. Создается отдельный набор данных(словарь стоп-слов), по которому должны проверить вхождение слов документа в словарь стоп-слов.

Выполним требования следующим образом:

1) получить по документу список существительных, глаголов и т.д

2) очистить от пунктуации

3) проверить на вхождение в словарь стоп-слов. Библиотека предоставляет ограниченный функционал, также стоп-слова для каждой предметной области и для каждой задачи могут быть специфическими - и могут быть определены пользователем самостоятельно.
   

Будем искать только существительные, а это значит, что междометия/предлоги/и т.д. по определению не попадут в выборку.
Также применим функцию:

.is_punct - функция поиска пунктуации

.is_stop - функция проверки на вхождение в словарь стоп-слов (необязательно)


```python
nouns = [ token.text for token in doc
                               if token.pos_ == 'NOUN' and not token.is_punct and not token.is_stop]
nouns
        
```

После получения существительных документа вычислите число употреблений слов:

Counter(...)

Далее примените к результату Counter(...) функцию .most_common(10)

```python
nouns_counter = Counter(nouns)
nouns_counter
```

```python
nouns_top10 = nouns_counter.most_common(10)
nouns_top10
```

```python
verbs = [token.text for token in doc
                                if token.pos_ == 'VERB' and not token.is_punct and not token.is_stop]
verbs_counter = Counter(verbs)
verbs_top10 = verbs_counter.most_common(10)
verbs_top10
```

```python
adjs = [token.text for token in doc
              if token.pos_ == 'ADJ' and not token.is_punct and not token.is_stop]
adjs_counter = Counter(adjs)
adjs_top10 = adjs_counter.most_common(10)
adjs_top10
```

```python
index = [word.text for word in doc]
columns = [word.text for word in doc]
```

```python
matrix = pd.DataFrame(index=index, columns=columns)
```
