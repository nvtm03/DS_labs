### 3.1 Задача регрессии на примере набора данных iris.csv

В прошлой лабораторной работы вы решали задачу классификации - вы разрабатывали модель и обучали её классифицировать по параметрам цветка его принадлежность к определённому виду(к определенной категории). Где:

'sepal_length','sepal_width', 'petal_length','petal_width' - признаки и независимые переменные.

'species'- искомая и зависимая переменная, категория.

В данной лабораторной работе необходимо решить задачу регресии - разработать модель и обучить её определять значение длины чашелистника цветка ('sepal_length') в зависимости от его категории 'species' и значений 'sepal_width', 'petal_length','petal_width.

Иначе, теперь
'sepal_width', 'petal_length','petal_width', 'species' - признаки и независимые переменные.

'sepal_length' - искомая и зависимая переменная.

```python
#Из библиотеки sklearn.model_selection импортирууйте train_test_split, данная функция помогает автоматизаровать разделение набора данных и произвольным образом выбрать строки из набора данных 
#см.предыдущую л.р.
from sklearn.model_selection import train_test_split
```

```python
# разделите набор данных в соотношении
# обучение - 70
# тестирование - 30
# теперь Y равно'sepal_length' (см.по аналогии как в предыдущ.л.р.)
Y = df['sepal_length']
X = df.drop(columns=['sepal_length'])


x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
#импорт линейной регрессии
from sklearn.linear_model import LinearRegression
```

```python
#Оставьте такое наименование вашей модели
model_lin_reg = LinearRegression()
```

```python
# Обучите модель
model_lin_reg.fit(x_train, y_train)
```

```python
# Вывесдите метрику, показывающую точность работы модели
print("Точность: ",model_lin_reg.score(x_test, y_test) * 100)
```

```python
#Новый набор данных для проверки работоспособности модели, задаём признаки цветка
data = {'sepal_width':[3.6], 'petal_length':[1.4],'petal_width': [0.2], 'species':[0]}
```

```python
#Создадим новый набор данных вручную
new_df = pd.DataFrame()
```

```python
#Экземпляр данных поместим в датафрейм
new_df = pd.DataFrame(data)
```

```python
#Примените predict - предсказание значения длины чашелистника цветка
predicted_with_lin_reg = model_lin_reg.predict(new_df)
```

```python
#Примените print для отображения результата
print(predicted_with_lin_reg)
```

```python
#Задайте новый набор данных и задайте в нем другие значения, чтобы теперь на нём проверить работу модели
#Примените predict
#Примените print для отображения результата
data_test = {'sepal_width':[2.8,3.4,3.9,4.2,4.1], 'petal_length':[2.8,3.5,3.9,4.1,4.1],'petal_width': [0.5,0.9,1.4,2.5,2.2], 'species':[0,1,1,2,0]}

print(model_lin_reg.predict(pd.DataFrame(data_test)))
```
